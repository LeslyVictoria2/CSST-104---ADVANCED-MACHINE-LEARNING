**"Exploratory Data Analysis of Product Data"**

**1 . Data Loading:**

```python
#Load the hardwareStore.csv dataset into a pandas DataFrame.
import pandas as pd
import matplotlib.pyplot as plt

dt = pd.read_csv('hardwareStore.csv')
```

**2. Initial Data Analysis:**

```python
#Display basic statistics of the dataset, including mean, median, and standard deviation of numeric columns.
statistics = dt.describe()
print(statistics, '\n')

#Identify the unique categories and their counts.
category = dt['CATEGORY_NAME'].value_counts()
print(category)
```

**3. Category Analysis:**

```python
#Create a bar chart to visualize the distribution of products across categories.
category.plot(kind = 'bar')
plt.xlabel('CATEGORY_NAME')
plt.ylabel('PRODUCT_ID')
plt.title('Distribution of Products across Categories')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Determine which category has the highest number of products.
highest_product = category.idxmax()
print('\nThe highest number of products: ' + highest_product)
```

**4. Cost and Price Analysis:**

```python
#Calculate the average standard cost and average list price.
average_of_sc = dt['STANDARD_COST'].mean()
print('Averange standard cost: {:.2f}'.format(average_of_sc))
average_of_lp = dt['LIST_PRICE'].mean()
print('Averange list price: {:.2f}'.format(average_of_lp))

#Identify the product with the highest standard cost and the one with the highest list price.
sc = dt.groupby('PRODUCT_NAME')['STANDARD_COST'].sum()
highest_sc = sc.idxmax()
s = dt.groupby('STANDARD_COST')['STANDARD_COST'].sum().idxmax()
print('\nHighest Standard Cost: ', highest_sc +' =', s)

lp = dt.groupby('PRODUCT_NAME')['LIST_PRICE'].sum()
highest_lp = lp.idxmax()
l = dt.groupby('LIST_PRICE')['LIST_PRICE'].sum().idxmax()
print('Highest List Price: ', highest_lp +' =', l)
```

**5. Location Analysis:**

```python
#Determine the location (city or state) with the highest quantity of products in stock.
location = dt.groupby('STATE')['QUANTITY'].sum()
highest_l = location.idxmax()
highest_q = dt.groupby('QUANTITY')['QUANTITY'].sum().idxmax()
print('Highest Location with the Highest Quantity of Products in Stock: ', highest_l +' =', highest_q)
```

**6. Insights and Reporting:**

```python
#Summarize your findings and insights from the data analysis.
#Create a report or presentation that includes visualizations and explanations.
report = """
Analysis Report

1.Data Loading:
  Dataset loaded successfully

2.Initial Data Analysis:
    Basic statistics of the Dataset:
    {}
    Unique categories and their counts:
    {}

3.Category Analysis:
    Bar chart displaying distribution of products across categories is attached
    Category with the highest number of products:
    {}

4.Cost and Price Analysis:
    Average standard cost: {: .2f}
    Average list price: {: .2f}
    Product with the highest standard cost: {}
    Product with the list price: {}

5.Location Analysis:
    Location with the highest quantity of products in stock: {}

6.Insights:
    The datasets contains a diverse range of products across various categories.

""". format(statistics, category, highest_product, average_of_sc,
            average_of_lp, highest_sc, highest_lp,  highest_l)

print(report)
with open ('analysis_report.txt', 'w') as f:
  f.write(report)
```
