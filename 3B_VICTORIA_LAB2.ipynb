**1. Creating Dataset:**

```python
import pandas as pd

data = {
    'student_id':range(1,16),
    'gender':["", 'Male', 'Female', 'Female', 'Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male', 'Male', 'Female', 'Male'],
    'test_date': pd.date_range(start='2024-02-14', periods=15),
    'study_hours': [2,6,2,9,4,3,2,8,3,1,7,2,4,9,7],
    'test_score': [84,76,83,99,86,81,84,98,80,78,88,82,85,"",94],
    'attendance': ['Present', 'Present', 'Absent', 'Present', 'Present', "",'Present', 'Present', 'Present', 'Present','Present', 'Present', 'Absent', 'Absent', 'Absent']
}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

**2. Data Loading and Initial Examination:**

```python
df = pd.read_csv('student_performance.csv')

print(df.head(15))

print(df.isnull().sum())
```

**3. Exploratory Data Analysis:**

```python
import matplotlib.pyplot as plt

import seaborn as sne

summary_state = df [['test_score', 'study_hours']].describe()

plt.figure(figsize=(12,6))

#Histogram
plt.subplot(1,2,1)
sne.histplot(df['test_score'], bins = 15, kde = True)
plt.title('Distribution of the test score')

#Histogram
plt.subplot(1,2,2)
sne.histplot(df['study_hours'], bins = 15, kde = True)
plt.title('Distribution of the study hours')

#Scatterplot
plt.figure(figsize=(8,6))
sne.scatterplot(data = df, x = 'study_hours', y = 'test_score')
plt.title('Scatter Plot of Study Hours vs. Test Score')
```

**4. Data Visualization:**

```python
#Gender Distribution
plt.figure(figsize=(8,4))
sne.countplot(data=df, x = 'gender')
plt.title('Gender Distribution')

#Attendance Distribution
plt.figure(figsize=(6,3))
sne.countplot(data=df, x = 'attendance')
plt.title('Attendance Distribution')

#Test score over time
plt.figure(figsize=(17,6))
sne.lineplot(data=df, x = 'test_date', y ='test_score')
plt.title('Tred in Test Score over time')

#Heatmap
correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()
plt.figure(figsize=(8,6))
sne.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```
