**"Simple Linear Regression"**

**1. Importing Libraries:**

```python
import pandas as pd #Data Manupilation
import numpy as np #Mathematical or Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Training Data Modeling
```

**2. Loading Dataset and Preparing Data:**

```python
dt = pd.read_csv('Lesson_4_student_grades_dataset.csv')

independent = dt['Hours_of_Study'].values.reshape(-1, 1)
dependent = dt['Grades']
```

**3. Visualizing the Data:**

```python
plt.scatter(independent,dependent)
plt.xlabel('Independent Variable (Hours of Study)')
plt.ylabel('Dependent Variable (Grades)')
plt.title('Scatter Plot of Data')
plt.show()
```

**4. Fitting the Linear Regression Model:**

```python
model = LinearRegression()
model.fit(independent, dependent)
```

**5. Getting Model Coefficients:**

```python
slp = model.coef_[0]
intercept = model.intercept_
print("THE SLOPE IS: ", slp)
print("THE INTERCEPT IS: ", intercept)
```

**6. Visualizing the Linear Regression Line:**

```python
plt.scatter(independent,dependent)
plt.plot(independent,model.predict(independent), color='darkred')
plt.xlabel('Independent Variable (Hours of Study)')
plt.ylabel('Dependent Variable (Grades)')
plt.title('Linear Regression')
plt.show()
```

**7. Making Predications:**

```python
new_independent = [[np.random.randint(1,12)]]
predicted_dependent = model.predict(new_independent)
print('VALUE OF NEW HOURS OF STUDY: ',new_independent)
print('PREDICTED GRADE: ',predicted_dependent)
```
